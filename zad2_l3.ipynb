## ZADANIE 2

Metodą aproksymacji średniokwadratowej (dyskretnej) wyznaczyć wielomian stopnia
co najwyżej: a) drugiego, b) czwartego, aproksymujący funkcję y = sin x, x ∈ [0, π/2]
przyjmując węzły aproksymacji w punktach 0, π/6, π/4, π/3, π/2. Sporządzić wykres
funkcji i wielomianu aproksymującego.

$ \sum_{k=0}^{n}a_{k} \sum_{i=1}^{m}x_{i}^{j+k} = \sum_{i=1}^{m}y_{i}x_{i}^{j}, for j = 0,1,...,n$

```python
def uklad_rownan(x, y, n):
    xs = np.array([]); xy = np.array([])
    for i in range(0, (n + 1)):
        for exp in range(0, (n + 1)):
            tx = np.sum(x**(i + exp))
            xs = np.append(xs, tx)
        ty = np.sum(y * (x**i))
        xy = np.append(xy, ty)
    return xs, xy

def func(x, a):
    px = 0
    for j in range(0, np.size(a)):
        px += (a[j] * (x**j))
    return px
```

```python
n = 2 #stopień
x = np.array([0, math.pi / 6, math.pi / 4, math.pi / 3, math.pi / 2])
y = np.array([0, 0.5, (2**0.5)/2, (3**0.5)/2, 1])

x_lin = np.linspace(x[0], x[-1], 1000) #zakres od 1 do ostatniego pkt-u
xs, xy = uklad_rownan(x, y, n) #xs - suma wartosci x , xy - iloczyn x i y
print(xs)
xs = np.reshape(xs, ((n + 1), (n + 1))) #zmiana wymiarow macierzy
print(xs)
#print(xs, xy)
xy = np.reshape(xy, ((n + 1), 1)) #zmiana wymiarów macierzy
a = np.linalg.solve(xs, xy)


#print(func(x_lin, a))
#print(a)
print("Wielomian aproksymujący przyjmuje postać określoną równaniem", str(a[0]), "x^2 +", str(a[1]), "x +", str(a[2])  )
```

```python
plt.scatter(x,y, label='wartości sin(x)')
plt.plot(x_lin, func(x_lin, a), label='aproksymacja', color = "red")
plt.title("Wykres aproksymujący")
plt.legend()
plt.grid()
```
