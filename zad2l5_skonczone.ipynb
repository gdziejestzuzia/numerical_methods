```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def rungeKutta(x0, y0, x, h) :
    
    f_poch = lambda x, y: (math.pi**2 * np.sin(math.pi * x)-(x+2)*(x - np.sin(math.pi * x)) + (x+2)*y )
    
    n = round((x - x0) / h)
    y = y0;

    for i in range(1, n + 1) :
        k1 = h * f_poch(x0, y);
        k2 = h * f_poch(x0 + h, y + k1);
        y = y + 0.5 * (k1 +  k2);
        
        k1 = h * y
        k2 = h * y
        y = y + 0.5 *(k1+k2)

        x0 = x0 + h;

    return y;
```

```python
xs = np.linspace(0,1, 100)
y_list =[]
y_2 = []
s = 0
h = 0.01
for xi in xs:
    y_i = rungeKutta(0, s, xi, h)
    y_list.append(y_i/100)
    
```

```python
f = lambda x: x - np.sin(math.pi * x) 
x = np.arange(0, 1+h, h) 
plt.figure(figsize = (12, 8))
plt.plot(xs, y_list, 'bo-', label='metoda R-K 2st.')
plt.plot(x, f(x), 'r', label='dokładne wartosci')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.legend(loc='lower right')
plt.title("Porównanie rozwiązań ")
plt.show()
```
