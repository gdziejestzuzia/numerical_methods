Uzasadnij, że złożona metoda trapezów zastosowana do obliczenia całek
$$ \int_{0}^{2\pi} cos(x) \,dx , \int_{0}^{2\pi} sin(x) \,dx $$
podaje dokładny wynik. Zilustruj to obliczeniami numerycznymi

$ \int_{0}^{2\pi} cos(x) \,dx = 0 $

```python
def trapez(f, a, b, n):
    h = float(b-a)/n
    result = 0.5 *( f(a) + f(b))
    for i in range(1, n):
        result += f(a + i * h)
    result *= h
    return result

v = lambda x: math.cos(x)
n = 10
rozwiazanie_numeryczne = trapez(v, 0, 2 * math.pi, n)

#v(2* math.pi) - v(0)
```
